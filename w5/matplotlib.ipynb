# Matplotlib Scatter

```python
x = np.array([0, 4, 2, 8, 1, 5, 5, 8, 9, 9, 2, 8, 4, 9, 3, 2, 6, 5, 4, 0])
y = np.array([2, 5, 8, 0, 2, 1, 7, 0, 2, 2, 0, 1, 2, 7, 8, 2, 6, 4, 3, 4])

plt.scatter(x, y)
plt.show()
```

## Compare Plots

```python
# Mid term score
x = np.array([0, 4, 2, 8, 1, 5, 5, 8, 9, 9, 2, 8, 4, 9, 3, 2, 6, 5, 4, 0])
y = np.array([2, 5, 8, 0, 2, 1, 7, 0, 2, 2, 0, 1, 2, 7, 8, 2, 6, 4, 3, 4])
plt.scatter(x, y)

# Final term score
x = np.array([4, 4, 7, 7, 9, 7, 9, 1, 3, 3, 4, 3, 6, 6, 5, 6, 0, 1, 5, 7])
y = np.array([1, 4, 3, 5, 0, 6, 0, 9, 5, 8, 8, 6, 1, 3, 3, 6, 0, 5, 8, 2])
plt.scatter(x, y)

plt.show()
```

## Colors

```python
# Mid term score
x = np.array([0, 4, 2, 8, 1, 5, 5, 8, 9, 9, 2, 8, 4, 9, 3, 2, 6, 5, 4, 0])
y = np.array([2, 5, 8, 0, 2, 1, 7, 0, 2, 2, 0, 1, 2, 7, 8, 2, 6, 4, 3, 4])
plt.scatter(x, y, color='r')

# Final term score
x = np.array([4, 4, 7, 7, 9, 7, 9, 1, 3, 3, 4, 3, 6, 6, 5, 6, 0, 1, 5, 7])
y = np.array([1, 4, 3, 5, 0, 6, 0, 9, 5, 8, 8, 6, 1, 3, 3, 6, 0, 5, 8, 2])
plt.scatter(x, y, color='g')

plt.show()
```

## Transparent

```python
x = np.array([0, 7, 7, 6, 8])
y = np.array([3, 2, 6, 6, 9])
sizes = np.array([10, 20, 300, 40, 90])

plt.scatter(x, y, s=sizes, cmap='flag', alpha=0.5)

plt.colorbar()
plt.show()
```

## Combine Color Size and Alpha

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Matplotlib Pie Charts

```python
y = np.array([5, 5, 8, 2, 7])

plt.pie(y)
plt.show() 
```

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']

plt.pie(y, labels=mylabels)
plt.show() 
```

## Start angle

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']

plt.pie(y, labels=mylabels, startangle=90)
plt.show() 
```

## Explode

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']
explodes = [0.2,0,0,0,0]

plt.pie(y, labels=mylabels, explode=explodes)
plt.show() 
```

## Shadow

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']
explodes = [0.2,0,0,0,0]

plt.pie(y, labels=mylabels, explode=explodes, shadow=True)
plt.show() 
```

## Colors

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']
explodes = [0.2,0,0,0,0]
mycolors = np.array(['red', 'blue', 'green', 'pink', 'orange'])

plt.pie(y, labels=mylabels, explode=explodes, shadow=True, colors=mycolors)
plt.show()
```

## Legend

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']
explodes = [0.2,0,0,0,0]
mycolors = np.array(['red', 'blue', 'green', 'pink', 'orange'])

plt.pie(y, labels=mylabels, explode=explodes, shadow=True, colors=mycolors)
plt.legend()
plt.show()
```

## Show values

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']
explodes = [0.2,0,0,0,0]
mycolors = np.array(['red', 'blue', 'green', 'pink', 'orange'])

plt.pie(y, labels=mylabels, explode=explodes, shadow=True, colors=mycolors, autopct='%1.1f%%')
# plt.legend()
plt.show()
```

```python
# Modified autopct: autopct = lambda p:'any text you want'

plt.pie(y, labels=mylabels, autopct=lambda p:f'{p:.2f}%, {p*sum(y)/100 :.0f} items')
plt.show()
```

# Matplotlib Save Fig

```python
y = np.array([5, 5, 8, 2, 7])
mylabels = ['A','B','C','D','E']
explodes = [0.2,0,0,0,0]
mycolors = np.array(['red', 'blue', 'green', 'pink', 'orange'])

fig = plt.figure()
plt.pie(y, labels=mylabels, explode=explodes, shadow=True, colors=mycolors, autopct='%1.1f%%')
# plt.legend()
plt.show()
```

## Working with text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
# n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
