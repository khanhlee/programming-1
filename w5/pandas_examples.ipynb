# Iterating over Rows and Columns in DataFrame

```python
df_renamed.plot()
plt.show()
```

```python
# iterrows()
for i, row in df_renamed.iterrows():
    print('Index: ', i)
    print(row)
```

```python
count = 0
for i, row in df_renamed.iterrows():
    print('Index: ', i)
    print(row)
    count += 1
    if count == 3:
        break
```

```python
# iteritems (for columns)
count = 0
for i, col in df_renamed.iteritems():
    print('Index: ', i)
    print(col)
    count += 1
    if count == 3:
        break
```

# Concatenating

```python
# concat()
df1 = df_renamed[['PREG','GLU','BP']]
df2 = df_renamed[['TC','INS']]
df3 = df_renamed[['BMI','AGE','DB']]
```

```python
df3
```

```python
frames = [df1,df2,df3]
new_df = pd.concat(frames, axis=1)
new_df
```

```python
df4 = df1.append(df3, ignore_index=True)
df4
```
