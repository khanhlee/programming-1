```python
import numpy as np
import matplotlib.pyplot as plt
raw_data = open('pima-indians-diabetes.csv')
data = np.loadtxt(raw_data, skiprows=1, delimiter=',')
data
```

## Correlation between Age and BMI

```python
data_bmi = data[:,5]
data_age = data[:,7]
```

```python
plt.plot(data_age, data_bmi, 'o', mec = '#50EF39', mfc = '#50EF39')
plt.show
```

### BMI Distribution

```python
plt.subplot(1, 2, 1)
plt.hist(data[:,5], color='#50EF39')
plt.title('BMI Distribution')

plt.subplot(1, 2, 2)
plt.hist(data[:,7], color='pink')
plt.title('Age Distribution')

plt.show()
```
